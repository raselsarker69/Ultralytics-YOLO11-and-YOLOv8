## Track

```python
#Bot Sort
results = model.track(source="/content/1.mp4", persist=True, save=True)
```

```python
# Byte Track
from ultralytics import YOLO

# Load the model
model = YOLO("yolo11n.pt")  # Specify your model file

# Perform object tracking with ByteTrack
results = model.track(
    source="/content/palace.mp4",  # Input video
    persist=True,  # Maintain object IDs
    show=True,  # Display the video with tracking in real-time
    classes=0,  # Track only a specific class (e.g., "person")
    tracker="bytetrack.yaml"  # Specify ByteTrack as the tracking algorithm
)
```
